#### Part 1.1 What is Keras?

**Keras:**
* Open-source library
* High-level deep learning framework
* Takes less code and enables fast experimentation
* It runs on top of other frameworks (like Tensorflow, Theano, or CNTK)

**Why choose Keras?**
* Fast industry-ready models
* Good for both beginners and experts
* Less code
* Build any architecture

**Deep Learning advantages:**

#### Part 1.2 Your first neural network

**Neural Network:**
* Fetch training data into **input layer**, predict a value at **output layer**
* Connection from neuron to another: **weight** and **bias weight**
* Output of neuron can also be transformed by **activation function**
* Tuning the weights of parameters is the key of learning in neural networks
* Common algorithm: Gradient Descent algorithm via a process known as back-propagation

**Two ways to build models in Keras:**
1. The Sequential API:
    * Model as a stack of layers
    * Start with input layer
    * Add a couple of hidden layers
    * End with output layer
2. Functional API (not covered here)

#### Part 2.1 Binary Classification

* Binary classification problems:
    * When you predict whether an observation belongs to one of two possible classes
* Seaborn's *'.pairplot()'* can be used to explore binary classification dataset
* Neuron networks: input layer, hidden layers, output layer
* **Sigmoid activation function** is common for output layer in binary classification problems
* **Sigmoid activation function** for the last output node, it converts the output to a floating number between **0 and 1**
* When compiling model, use *loss='binary_crossentropy'* as loss function
* Then, we can train our model and use to make predictions

**Alternative ways:** (we used this in Course 15: introduction to Deep Learning)
* **Softmax activation function** for the output layer, set the number of nodes to 2
* When compiling model, use *loss='categorical_crossentropy'* as loss function

```python
# Code for part 2.1

# exploring dollar bills
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# get dataset
banknotes = pd.read_csv('16_datasets/banknotes.csv')

# use pairplot to explore dataset
sns.pairplot(banknotes, hue='class')
plt.show()

# describe dataset
print('Dataset stats: \n', banknotes.describe())
print('Observation per class: \n', banknotes['class'].value_counts())
```

```python
# Code for part 2.1 (continue)

# a binary classification model
from keras.models import Sequential
from keras.layers import Dense

# create the model
model = Sequential()
model.add(Dense(1, input_shape=(4,), activation='sigmoid'))

# compile the model
model.compile(optimizer='sgd', loss='binary_crossentropy', metrics=['accuracy'])
model.summary()
```

#### Part 2.2 Multi-class classification

* Multi-class classification problems:
    * Has more than 2 classes as outcome
    * Need to make a minor tweak in the neuron network architecture
* Neuron networks: input layer, hidden layers, output layer (same as binary classification problem)
* Differences:
    * Output layer: Has more than 1 nodes (1 node per class), use **'softmax' activation function** in output layer
    * Compiling model: Use **'categorical_crossentropy'** as loss function
    * Preparing data: Turn target column from one column to **one-hot encoded vector**:
        1. Use *'pd.get_dummies()'* or
        2. Use *'tf.keras.utils.to_categorical()'*

```python
# Code for part 2.2

# a multi-class model
from keras.models import Sequential
from keras.layers import Dense

# create the model
model = Sequential()
model.add(Dense(128, activation='relu', input_shape=(2,)))
model.add(Dense(64, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(4, activation='softmax'))

# compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

    * Output layer: *many* neurons (one neuron per class), but use *sigmoid activation* (value range 0-1)
    * Loss function: *'binary_crossentropy'* (as if performing several binary classification problems)

```python
# Code for part 2.3 

# an irrigation problem

from keras.models import Sequential
from keras.layers import Dense

# create the model
model = Sequential()
model.add(Dense(64, input_shape=(20,), activation='relu')) # hidden layer
model.add(Dense(3, activation='sigmoid')) # output layer

# compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.summary()
```

```python
# training with multiple labels
import pandas as pd
import tensorflow as tf
from sklearn.model_selection import train_test_split
import numpy as np

# get data
irrigation = pd.read_csv('16_datasets/irrigation_machine.csv')
# print(irrigation.shape)
sensors = irrigation.iloc[:,1:21]
parcels = irrigation.iloc[:, 21:24]
print(sensors.head())
print(parcels.head())

# split training and testing set
X_train, X_test, y_train, y_test = train_test_split(sensors, parcels, test_size=0.3)

# convert to numpy
X_train, X_test = X_train.to_numpy(), X_test.to_numpy()
y_train, y_test = y_train.to_numpy(), y_test.to_numpy()

# convert to tensor
X_train, X_test = tf.convert_to_tensor(X_train), tf.convert_to_tensor(X_test)
y_train, y_test = tf.convert_to_tensor(y_train), tf.convert_to_tensor(y_test)

# train the model
model.fit(X_train, y_train, epochs=100, validation_split=0.2)

# make predictions
preds = model.predict(X_test)
preds_rounded = np.round(preds)
print('Rounded Predictions: \n', preds_rounded)

# evaluate the model
accuracy = model.evaluate(X_test, y_test)[1]
print('Accuracy: ', accuracy)
```
