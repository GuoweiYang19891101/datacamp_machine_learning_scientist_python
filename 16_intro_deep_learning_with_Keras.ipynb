#### Part 1.1 What is Keras?

**Keras:**
* Open-source library
* High-level deep learning framework
* Takes less code and enables fast experimentation
* It runs on top of other frameworks (like Tensorflow, Theano, or CNTK)

**Why choose Keras?**
* Fast industry-ready models
* Good for both beginners and experts
* Less code
* Build any architecture

**Deep Learning advantages:**

#### Part 1.2 Your first neural network

**Neural Network:**
* Fetch training data into **input layer**, predict a value at **output layer**
* Connection from neuron to another: **weight** and **bias weight**
* Output of neuron can also be transformed by **activation function**
* Tuning the weights of parameters is the key of learning in neural networks
* Common algorithm: Gradient Descent algorithm via a process known as back-propagation

**Two ways to build models in Keras:**
1. The Sequential API:
    * Model as a stack of layers
    * Start with input layer
    * Add a couple of hidden layers
    * End with output layer
2. Functional API (not covered here)

#### Part 2.1 Binary Classification

* Binary classification problems:
    * When you predict whether an observation belongs to one of two possible classes
* Seaborn's *'.pairplot()'* can be used to explore binary classification dataset
* Neuron networks: input layer, hidden layers, output layer
* **Sigmoid activation function** is common for output layer in binary classification problems
* **Sigmoid activation function** for the last output node, it converts the output to a floating number between **0 and 1**
* When compiling model, use *loss='binary_crossentropy'* as loss function
* Then, we can train our model and use to make predictions

**Alternative ways:** (we used this in Course 15: introduction to Deep Learning)
* **Softmax activation function** for the output layer, set the number of nodes to 2
* When compiling model, use *loss='categorical_crossentropy'* as loss function

```python
# Code for part 2.1

# exploring dollar bills
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# get dataset
banknotes = pd.read_csv('16_datasets/banknotes.csv')

# use pairplot to explore dataset
sns.pairplot(banknotes, hue='class')
plt.show()

# describe dataset
print('Dataset stats: \n', banknotes.describe())
print('Observation per class: \n', banknotes['class'].value_counts())
```

```python
# Code for part 2.1 (continue)

# a binary classification model
from keras.models import Sequential
from keras.layers import Dense

# create the model
model = Sequential()
model.add(Dense(1, input_shape=(4,), activation='sigmoid'))

# compile the model
model.compile(optimizer='sgd', loss='binary_crossentropy', metrics=['accuracy'])
model.summary()
```

#### Part 2.2 Multi-class classification

* Multi-class classification problems:
    * Has more than 2 classes as outcome
    * Need to make a minor tweak in the neuron network architecture
* Neuron networks: input layer, hidden layers, output layer (same as binary classification problem)
* Differences:
    * Output layer: Has more than 1 nodes (1 node per class), use **'softmax' activation function** in output layer
    * Compiling model: Use **'categorical_crossentropy'** as loss function
    * Preparing data: Turn target column from one column to **one-hot encoded vector**:
        1. Use *'pd.get_dummies()'* or
        2. Use *'tf.keras.utils.to_categorical()'*

```python
# Code for part 2.2

# a multi-class model
from keras.models import Sequential
from keras.layers import Dense

# create the model
model = Sequential()
model.add(Dense(128, activation='relu', input_shape=(2,)))
model.add(Dense(64, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(4, activation='softmax'))

# compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

    * Output layer: *many* neurons (one neuron per class), but use *sigmoid activation* (value range 0-1)
    * Loss function: *'binary_crossentropy'* (as if performing several binary classification problems)

#### Part 2.4 Keras callbacks (early stopping)

**Callback:**
* A function executed after some other function, event, or task has finished (like a trigger)
* Every time you call the *'.fit()'* method, a callback object returned, you can check its history
* **Early stopping** is a good callback function that prevents over-fitting
* **Model checkpoint** is another good callback function to identify and save the best model

```python
# Code for part 2.4 

# the history callback
import pandas as pd
from sklearn.model_selection import train_test_split
import tensorflow as tf
from keras.models import Sequential
from keras.layers import Dense

# get data (use same data as above exercise)
irrigation = pd.read_csv('16_datasets/irrigation_machine.csv')
sensors = irrigation.iloc[:, 1:21]
parcels = irrigation.iloc[:, 21:24]

# split into training and testing set
X_train, X_test, y_train, y_test = train_test_split(sensors, parcels, test_size=0.3)
X_train, X_test = X_train.to_numpy(), X_test.to_numpy()
y_train, y_test = y_train.to_numpy(), y_test.to_numpy()
X_train, X_test = tf.convert_to_tensor(X_train), tf.convert_to_tensor(X_test)
y_train, y_test = tf.convert_to_tensor(y_train), tf.convert_to_tensor(y_test)

# create the model
model = Sequential()
model.add(Dense(64, input_shape=(20,), activation='relu')) # hidden layer
model.add(Dense(3, activation='sigmoid')) # output layer

# compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.summary()

# train the model and save history
h_callback = model.fit(X_train, y_train, epochs=50, validation_data=(X_test, y_test))

# create functions to plot loss and accuracy
def plot_loss(loss,val_loss):
    plt.figure()
    plt.plot(loss)
    plt.plot(val_loss)
    plt.title('Model loss')
    plt.ylabel('Loss')
    plt.xlabel('Epoch')
    plt.legend(['Train', 'Test'], loc='upper right')
    plt.show()

def plot_accuracy(acc,val_acc):
    # Plot training & validation accuracy values
    plt.figure()
    plt.plot(acc)
    plt.plot(val_acc)
    plt.title('Model accuracy')
    plt.ylabel('Accuracy')
    plt.xlabel('Epoch')
    plt.legend(['Train', 'Test'], loc='upper left')
    plt.show()

# plot train and test loss during training
plot_loss(h_callback.history['loss'], h_callback.history['val_loss'])
```

#### Part 3.2 Activation functions

* Input of activation function:
    * Sum of inputs * weights + bias
$$a = x_{1} * w_{1} + x_{2} * w_{2} + x_{3} * w_{3} + b_{1}$$ 
* Output of activation function:
    * Transformed value from activation function
$$y = f(a)$$
* Examples:
    * 'tanh'
    * 'ReLU'
    * 'sigmoid'
    * 'leaky_ReLU'

**Which activation function to use?**
* No magic formula
* Different properties
* Depends on our problem
* Goal to achieve in a given layer
* ReLU are a good first choice
* Sigmoid are not recommended for deep models
* Tune with experimentation

#### Part 3.3 Batch size and batch normalization

* A mini-batch is a subset of data samples
* We DO NOT update network weights only ONCE per epoch
* We take several updates, one per mini-batch
* Set *'batch_size'* parameter in *'model.fit()'* method

**Advantages of mini-batch:**
* Networks train faster (more weight updates in same amount of time)
* Less RAM memory required, can train on huge datasets
* Noise by small batch-size can help escape local minimal

**Disadvantages of mini-batch:**
* More iterations need to be run
* Need to find a good batch size

**Batch Normalization:**
* Normalization is a common pre-processing step in machine learning algorithms
* Improves gradient flow
* Allows higher learning rates
* Reduces dependence on weight initializations
* Acts as an unintended of regularization
* Limits internal co-variate shift
* It is applied between two layers (mostly after input layer)

#### Part 3.4 Hyperparameter tuning

**Neural network hyperparameters:**
* Number of layers
* Number of neurons per layer
* Layer order
* Layer activations
* Batch sizes
* Learning rates
* Optimizers
* And so on....

**How to tune these?**
* In *'sklearn'*, we use *'RandomizedSearchCV()'* method to achieve this
* For neural networks:
    1. Transform model into sklearn estimators with *'KerasClassifier()'* method
    2. Perform random search on the model
    3. Perform cross-validation on estimator with *best parameters* 
       
    
**Tips:**
* Random search is preferred over grid search
* Don't use many epochs
* Use a smaller sample of your dataset
* Play with batch sizes, activations, optimizers and learning rates

```python
# Code for part 3.4

# preparing a model for tuning
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler

# define function to create model
def create_model(learning_rate=0.01, activation='relu'):
    opt = Adam(learning_rate=learning_rate)
    
    model = Sequential()
    model.add(Dense(128, input_shape=(30,), activation=activation))
    model.add(Dense(256, activation=activation))
    model.add(Dense(1, activation='sigmoid'))
    
    model.compile(optimizer=opt, loss='binary_crossentropy', metrics=['accuracy'])
    return model

# get data
breast = pd.read_csv('16_datasets/breast_cancer.csv')
X = breast.drop(['id','diagnosis', 'Unnamed: 32'], axis=1)
y = (breast['diagnosis']=='M').astype(int)
X, y = X.to_numpy(dtype=np.float32), y.to_numpy(dtype=np.float32)

# standardize features
scaler = StandardScaler()
X = scaler.fit_transform(X)
```

```python
# Code for 3.4 (continue)  

# tuning the model parameters
from scikeras.wrappers import KerasClassifier
from sklearn.model_selection import RandomizedSearchCV, KFold


# create a classifier
model = KerasClassifier(build_fn=create_model)

# define params
params = {
          # 'activation': ['relu', 'tanh'],
          'batch_size': [32, 128, 256],
          'epochs':[50, 100, 200],
          # 'learning_rate': [0.1, 0.01, 0.001]
          }

# create a random search
random_search = RandomizedSearchCV(estimator=model, param_distributions=params, cv=KFold(n_splits=3))

# find the best parameters
random_search.fit(X, y)
print('Best parameters: ', random_search.best_params_)  
```

```python
# Code for 3.4 (continue)

# training with cross-validation
from scikeras.wrappers import KerasClassifier
from sklearn.model_selection import cross_val_score

# create a classifier with the best parameters (
model = KerasClassifier(build_fn=create_model(learning_rate=0.001, activation='relu'),
                        epochs=50, batch_size=128, verbose=0)

# calculate the accuracy score
kfolds = cross_val_score(model, X, y, cv=3)
print('The mean accuracy was: ', kfolds.mean())
print('With a standard deviation of: ', kfolds.std())
```

**Convolutional Neural Networks(CNNs):**
* A different type of network that used in *computer visions* and other areas
* It uses convolutional layers which uses a *mathematical operation* to preserve spatial relationships
* It applies a **filter, or kernel**, of a given size as we slide through each pixel in the image
* Generates a new image where certain characteristics are amplified

```python
# Code for part 4.2 

# building a CNN model
from keras.models import Sequential
from keras.layers import Conv2D, Flatten

# build the model for MNIST dataset
model = Sequential()
# add a convolutional layer of 32 filters of size 3x3
model.add(Conv2D(filters=32, kernel_size=3, input_shape=(28, 28, 1), activation='relu')) 
# add a convolutional layer of 16 filters of size 3*3
model.add(Conv2D(16, kernel_size=3, activation='relu'))
# flatten output from previous layers
model.add(Flatten())
# add output layer
model.add(Dense(10, activation='softmax'))
```

#### Part 4.3 Intro to LSTMs

**Long Short Term Memory networks (LSTMs):**
* It is a type of recurrent neural networks (RNNs) which use past predictions in order to infer new ones
* Its units are called units or cells
* There is an internal state that is passed between units (memory of past steps)
* A unit receives: 
    1. Internal state of previous unit
    2. Output of previous unit
    3. A new input at time $t$
* Then, it updates:
    1. Updated state for next unit
    2. New output returned of LSTM unit
    3. Input to next LSTM unit
      
**When to use LSTMs?**
* Image captioning
* Speech to text
* Text translation
* Document summarization
* Text generation
* Musical composition
* And so on...

```python
# Code for part 4.3 

# text prediction with LSTMs
from keras.preprocessing.text import Tokenizer

# get text
text = 'it is not the strength of the body but the strength of the spirit it is useless to meet revenge with revenge it will heal nothing even the smallest person can change the course of history all we have to decide is what to do with the time that is given us the burned hand teaches best after that advice about fire goes to the heart'

# split text into array of words
words = text.split()

# make sentences of 4 words each
sentences = []
for i in range(4, len(words)):
    sentences.append(' '.join(words[i-4:i]))
    
# instantiate a tokenizer and fit it to sentences
tokenizer = Tokenizer()
tokenizer.fit_on_texts(sentences)

# turn sentences into numbers
sequences = tokenizer.texts_to_sequences(sentences)
print('Sentences: \n {} \n Sequences: {}'.format(sentences[:5], sequences[:5]))
```

```python
# Code for part 4.3 (continue)

# build your LSTM model
from keras.models import Sequential
from keras.layers import Embedding, LSTM, Dense

# get vocab size
vocab_size = 44

# build the model
model = Sequential()
model.add(Embedding(input_dim=vocab_size, input_length=3, output_dim=8,))
model.add(LSTM(32))
model.add(Dense(32, activation='relu'))
model.add(Dense(vocab_size, activation='softmax'))
model.summary()
```

```python
# Code for part 4.3 (continue)

# decode your predictions
import numpy as np

# create function to predict text
def predict_text(test_text, model=model):
    if len(test_text.split()) !=3:
        print('Text input should be 3 words!')
        return False
    
    # turn the test_text into a sequence of numbers
    test_seq = tokenizer.texts_to_sequences([test_text])
    test_seq = np.array(test_seq)

    # use the model to predict next word and return the result
    pred = model.predict(test_seq).argmax(axis=1)[0]
    return tokenizer.index_word[pred]

# test function
print(predict_text('meet revenge with'))
print(predict_text('the course of'))
print(predict_text('strength of the'))
```
