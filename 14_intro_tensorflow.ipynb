### CHAPTER 1. Introduction to TensorFlow

#### Part 1.1 Constants and variables
* TensorFlow is an open-source library for graph-based numerical computation
* Tensor is described as a generalization of vectors and matrices
*  You can think of it as a collection of numbers into a specific shape

**Constants in TensorFlow:**
* A constant is the simplest category of tensor
* Does not change and cannot be trained
* Can have any dimension

**variables in TensorFlow:**
* A variable's value can change during computation
* The value is shared, persistent, and modifiable
* The data type and shape are fixed

#### Part 1.2 Basic operations

* TensorFlow graph contains edges and nodes, edges are tensors and nodes are operations
* Perform add operation with *'add()'*:
    * Tensors should have same shape
* Perform element-wise multiplication operation with *'multiply()'*:
    * Tensors should have same shape
    * Example: [1,2,3] and [3,4,5] or [1,2] and [3,4]
* Perform matrix multiplication operation with *'matmul()'*:
    * Multiply A by B with *'matmul(A, B)'* 
    * Columns of A equal rows of B
    * Shape of result has rows of A and columns of B
* Perform summing over tensor dimension with *'reduce_sum()'*:
    * Sums over all dimension of tensor A with *'reduce_sum(A)'*
    * Sums over dimension i with *'reduce_sum(A, i)'*

```python
# Code for part 1.2 

# performing element-wise multiplication
from tensorflow import multiply, constant, ones_like

# define tensor A1 and A23 as constants
A1 = constant([1, 2, 3, 4])
A23 = constant([[1, 2, 3], [1, 6, 4]])

# define B1 and B23 to have the correct shape
B1 = ones_like(A1)
B23 = ones_like(A23)

# perform element-wise multiplication
C1 = multiply(A1, B1)
C23 = multiply(A23, B23)

# print the tensors C1 and C23
print('\n C1: {}'.format(C1.numpy()))
print('\n C23: {}'.format(C23.numpy()))
```

```python
# Code for part 1.2 (continue)

# making predictions with matrix multiplication
from tensorflow import constant, matmul

# define features, params, and bill as constants
features = constant([[2, 24], [2, 26], [2, 57], [1, 37]])
params = constant([[1000], [150]])
bill = constant([[3913], [2682], [8617], [64400]])

# compute billpred using features and params
billpred = matmul(features, params)
# print(billpred)

# compute and print the error
error = bill - billpred
print(error.numpy)
```

```python
# summing over tensor dimensions
from tensorflow import constant, reduce_sum

# get matrix 'wealth'
wealth = constant([[11, 50], [7, 2], [4, 60], [3, 0], [25, 10]])

# get summing over tensor dimension with 'reduce_sum()'
sum_0 = reduce_sum(wealth, 0).numpy()
sum_1 = reduce_sum(wealth, 1).numpy()
sum_2 = reduce_sum(wealth).numpy()

print(sum_0)
print(sum_1)
print(sum_2)
```

```python
# Code for part 1.3 (continue)

# optimizing with gradients
from tensorflow import Variable, GradientTape

# create a function to compute gradient/ slope by hand with loss function: y = x^2
def compute_gradient(x0):
    x = Variable(x0)
    with GradientTape() as tape:
        tape.watch(x)
        y = x * x
    return tape.gradient(y, x).numpy()

# try function on x = -1, 1, and 0
print(compute_gradient(-1.0))
print(compute_gradient(1.0))
print(compute_gradient(0.0))
```

```python
# Code for part 1.3 (continue)

# working with image data
import numpy as np
from tensorflow import reshape, matmul, reduce_sum

# create two tensors: letter, model
# model is used to classify letter into X or K
letter = np.array([[1., 0., 1.],
                   [1., 1., 0.],
                   [1., 0., 1.]])
model = np.array([[1., 0., -1.]])

# reshape tensor model to 3*1 vector
model = reshape(model, (3, 1))

# multiply letter by model
output = matmul(letter, model)

# sum over output and predict using numpy method
prediction = reduce_sum(output)
print(prediction.numpy())
```
